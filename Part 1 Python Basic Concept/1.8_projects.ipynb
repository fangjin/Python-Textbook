```python
## P1
pet_name = 'Fluffy'
hunger = 5  # Scale of 1-10, where 10 is very hungry
energy = 5  # Scale of 1-10, where 10 is full of energy
happiness = 5  # Scale of 1-10, where 10 is very happy

def status(evolution = 0):
    print(f"\n{pet_name}'s status: Hunger = {hunger}, Energy = {energy}, Happiness = {happiness}")
    if evolution and hunger >= 10 and energy >= 10 and happiness >= 10:
      print(f"\n{pet_name} has successfully evolved.")
      print(f"1.You need to upgrade your code yourself to complete the subsequent evolution design or even battles.")
      print(f"2.You can design or upload a character for your companions.")
      print(f"For more information, please contact your teacher.")
      return False
    else:
      return True

def feed():
    global hunger, happiness
    if hunger < 9:
        hunger += 2  # Feeding decreases hunger
        happiness += 1
        print(f"\nYou fed {pet_name}. Yum!")
    else:
        print(f"\n{pet_name} is not very hungry right now.")
    status()
    return True

def play():
    global happiness, energy, hunger
    if energy > 2:
        if happiness < 9:
          happiness += 2  # Playing increases happiness
        else:
          happiness = 10
        energy -= 2  # Playing decreases energy
        hunger -= 1
        print(f"\nYou played with {pet_name}. Fun!")
    else:
        print(f"\n{pet_name} is too tired to play.")
    status()
    return True

def sleep():
    global energy, happiness, hunger
    if energy < 8:
        energy += 3  # Sleeping increases energy
        hunger -= 1
        happiness -= 1
        print(f"\n{pet_name} is sleeping. Zzz...")
    else:
        print(f"\n{pet_name} is not very tired, {pet_name} do not want to sleep.")
    status()
    return True

# Main program loop
pet_name = input('Please give your pet a name.')
loop = True
while loop:
    choice = input(f"\nWhat would you like to do with {pet_name}? \n(Hint: You can type in one of your options: [status(),feed(),play(),sleep(),quit()])\n")

    # if choice == "status()":
    #     status()
    # elif choice == "feed()":
    #     feed()
    # elif choice == "play()":
    #     play()
    # elif choice == "sleep()":
    #     sleep()
    if choice == "quit()":
        print(f"\nGoodbye! {pet_name} will miss you!")
        break  # Exit the loop and end the program
    else:
        try:
          loop = eval(choice)
          # loop = exec(choice)
          # print(loop)
        except:
          print('You have type an error execution, please try again.')
```

```python
## P2.
def print_board(board):
    separator = " +---+---+---+"
    print("   0   1   2")
    print(separator)
    for idx, row in enumerate(board):
        print(f"{idx}| {' | '.join(row)} |")
        print(separator)
    print()

def check_winner(board, player):
    for i in range(3):
        if all(board[i][j] == player for j in range(3)) or \
           all(board[j][i] == player for j in range(3)):
            return True
    if all(board[i][i] == player for i in range(3)) or \
       all(board[i][2 - i] == player for i in range(3)):
        return True
    return False

def main():
    board = [[" " for _ in range(3)] for _ in range(3)]
    current_player = "X"
    while True:
        print_board(board)
        move = input(f"Player {current_player}, enter row and column (e.g., '1 2'): ")
        try:
            row, col = map(int, move.split())
        except ValueError:
            print("Invalid input. Please enter two integers separated by a space.")
            continue
        if not (0 <= row <= 2 and 0 <= col <= 2):
            print("Invalid input. Row and column must be between 0 and 2.")
            continue
        if board[row][col] != " ":
            print("That position is already taken!")
            continue
        board[row][col] = current_player
        if check_winner(board, current_player):
            print_board(board)
            print(f"Player {current_player} wins!")
            break
        if all(board[i][j] != " " for i in range(3) for j in range(3)):
            print_board(board)
            print("It's a tie!")
            break
        current_player = "O" if current_player == "X" else "X"

if __name__ == "__main__":
    main()
```

```python
## P3.
#simulate an ATM machine
account_number = "123456"
pin = "1234"
balance = 1000  # Starting balance

def login():
    user_account = input("Enter your account number: ")
    user_pin = input("Enter your PIN: ")
    if user_account == account_number and user_pin == pin:
        return True
    else:
        return False

def deposit():
    global balance
    amount = float(input("Enter amount to deposit: "))
    balance += amount
    print(f"${amount} deposited successfully.")

def withdraw():
    global balance
    amount = float(input("Enter amount to withdraw: "))
    if amount <= balance:
        balance -= amount
        print(f"${amount} withdrawn successfully.")
    else:
        print("Insufficient balance.")

def check_balance():
    print(f"Your current balance is: ${balance}")

def main():
    if login():
        print("Login successful!")
        while True:
            print("\n1. Deposit")
            print("2. Withdraw")
            print("3. Check Balance")
            print("4. Exit")
            choice = input("Choose an option (1-4): ")

            if choice == "1":
                deposit()
            elif choice == "2":
                withdraw()
            elif choice == "3":
                check_balance()
            elif choice == "4":
                print("Thank you for using the Mini ATM Simulator. Goodbye!")
                break
            else:
                print("Invalid choice. Please enter a number between 1 and 4.")
    else:
        print("Invalid account number or PIN.")

if __name__ == "__main__":
    main()
```

```python
## P4.
def print_grid(size, player_x, player_y):
    for x in range(size):
        for y in range(size):
            if x == player_x and y == player_y:
                print("P ", end="")  # Print P for Player's current position
            else:
                print(". ", end="")  # Print . for empty space
        print()  # New line at the end of each row

def play_game(size=5):
    player_x, player_y = 0, 0  # Starting position of the player

    while True:
        print("\nCurrent map:")
        print_grid(size, player_x, player_y)  # Show the grid

        move = input("Move (WASD): ").upper()
        if move == "W" and player_x > 0:
            player_x -= 1
        elif move == "S" and player_x < size - 1:
            player_x += 1
        elif move == "A" and player_y > 0:
            player_y -= 1
        elif move == "D" and player_y < size - 1:
            player_y += 1
        else:
            print("Invalid move or boundary reached. Please try again.")

if __name__ == "__main__":
    print("Navigate inside the grid. Use WASD keys to move: W (up), A (left), S (down), D (right).")
    play_game()
```
