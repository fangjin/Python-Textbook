```python
## P3.
# Simulation of an ATM machine
# Let us suppose there are several customers, we want to initialize their account and let one of them input and check his account.
#1. Initialize the account.
account_number = [123, 233, 456]
pin = [11, 22, 33]
balance = [0, 100, 1000]


# 2. Get some variables from input
user_account = input('Please enter your account.')
user_pin = input('Please enter your pin number.')
amount = float(input('Please enter the amount you want to withdraw.'))


# 3. Print out if the account_number is equal to user_account, the opposite to pin, and if amount is larger than your balance
print( user_account == account_number[0])
print( user_pin != pin[0])
print( amount > balance[0])

print( user_account == account_number[1])
print( user_pin != pin[1])
print( amount > balance[1])

print( user_account == account_number[2])
print( user_pin != pin[2])
print( amount > balance[2])

# brief introduction to numpy
import numpy as np
N_account = np.array(account_number)
N_pin = np.array(pin)
N_balance = np.array(balance)

print( N_account == user_account)
print( N_pin != user_pin)
print( N_balance > amount)
```
