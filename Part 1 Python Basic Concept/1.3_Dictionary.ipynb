```python
## P1
# Imagine that you've many pets (may be Pok√©mon?), and you plan to train them. Please first define and initial your pets. Hint(Let every variable be a list, and pets be a dictionary.)
pet_name = ['A','B','C']
hunger = [5,5,5]
energy = [6,7,4]
happiness = [2,7,10]

Pets = {'pet name':pet_name,'hunger':hunger,'energy':energy,'happiness':happiness}
print(Pets)

# How about let every pet be a dictionary?
A = {'pet name':'A','hunger':5,'energy':6,'happiness':2}
B = {'pet name':'B','hunger':5,'energy':7,'happiness':7}
C = {'pet name':'C','hunger':5,'energy':4,'happiness':10}

Pets = {'A':A,'B':B,'C':C}
print(Pets['A'])
print(Pets['B'])
print(Pets['C'])
```
