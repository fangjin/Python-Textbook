```python
#Q1
#Reference code
# Air quality dataset
# Academic dataset for creating bar charts
departments = ['Engineering', 'Humanities', 'Sciences', 'Arts', 'Business']
student_counts = [1200, 800, 1300, 700, 900]
graduate_students = [300, 150, 350, 200, 250]
undergraduate_students = [900, 650, 950, 500, 650]

#Answer code
import matplotlib.pyplot as plt


# Creating a basic bar chart
plt.figure(figsize=(10, 6))
plt.bar(departments, student_counts, color='skyblue')
plt.title('Total Students by Department')
plt.xlabel('Departments')
plt.ylabel('Number of Students')
plt.show()




```

```python
#Q3
#Reference code
# Air quality dataset
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'PM10': [20, 30, 45, 22, 29],  # PM10 levels
    'NO2': [30, 45, 22, 35, 40]    # NO2 levels
}

#Answer code
# Customized bar chart
plt.bar(departments, student_counts, width=0.6, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'],
        edgecolor='black', linewidth=1, tick_label=['Eng', 'Hum', 'Sci', 'Art', 'Bus'],
        align='center', label='Students 2020', hatch='/', alpha=0.7)
plt.title('Customized Student Enrollment Chart')
plt.xlabel('Department')
plt.ylabel('Total Students')
plt.legend(title='Year')
plt.show()
```
