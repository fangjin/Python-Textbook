```python
## P1
class Monster:
    def __init__(self, name):
        self.monster_name = name
        self.hunger = 5  # Scale of 1-10, where 10 is very hungry
        self.energy = 5  # Scale of 1-10, where 10 is full of energy
        self.happiness = 5  # Scale of 1-10, where 10 is very happy

    def status(self, evolution=0):
        print(f"\n{self.monster_name}'s status: Hunger = {self.hunger}, Energy = {self.energy}, Happiness = {self.happiness}")
        if evolution and self.hunger >= 10 and self.energy >= 10 and self.happiness >= 10:
            print(f"\n{self.monster_name} has successfully evolved.")
            print(f"1. You need to upgrade your code yourself to complete the subsequent evolution design or even battles.")
            print(f"2. You can design or upload a character for your companions.")
            print(f"For more information, please contact your teacher.")
            return False
        else:
            return True

    def feed(self):
        if self.hunger < 9:
            self.hunger += 2  # Feeding decreases hunger
            self.happiness += 1
            print(f"\nYou fed {self.monster_name}. Yum!")
        else:
            print(f"\n{self.monster_name} is not very hungry right now.")
        self.status()
        return True

    def play(self):
        if self.energy > 2:
            if self.happiness < 9:
                self.happiness += 2  # Playing increases happiness
            else:
                self.happiness = 10
            self.energy -= 2  # Playing decreases energy
            self.hunger -= 1
            print(f"\nYou played with {self.monster_name}. Fun!")
        else:
            print(f"\n{self.monster_name} is too tired to play.")
        self.status()
        return True

    def sleep(self):
        if self.energy < 8:
            self.energy += 3  # Sleeping increases energy
            self.hunger -= 1
            self.happiness -= 1
            print(f"\n{self.monster_name} is sleeping. Zzz...")
        else:
            print(f"\n{self.monster_name} is not very tired, {self.monster_name} do not want to sleep.")
        self.status()
        return True

def main():
    loop = True
    while loop:
        choice = input(f"\nWhat would you like to do with {name}? \n(Hint: You can type in one of your options: [status(),feed(),play(),sleep(),quit()])\n")
        if choice == "quit()":
            print(f"\nGoodbye! {name} will miss you!")
            break  # Exit the loop and end the program
        else:
            try:
                loop = eval(f"monster.{choice}")
            except:
                print('You have typed an error execution, please try again.')

name = input('Please give your monster a name: ')
monster = Monster(name)

if __name__ == '__main__':
    main()
```

```python
## Combat System:
import random

class CombatMonster(Monster):
    def __init__(self, name):
        super().__init__(name)
        self.health = 100  # Combat health points

    def attack(self, other_monster):
        damage = random.randint(5, 15)  # Random damage between 5 to 15
        other_monster.health -= damage
        print(f"\n{self.monster_name} attacks {other_monster.monster_name} and deals {damage} damage.")
        if other_monster.health <= 0:
            print(f"\n{other_monster.monster_name} has been defeated!")
        else:
            print(f"\n{other_monster.monster_name}'s health: {other_monster.health}")

def main():
    name = input('Please give your monster a name: ')
    monster = CombatMonster(name)

    loop = True
    while loop:
        choice = input(f"\nWhat would you like to do with {monster.monster_name}? \n(Hint: You can type in one of your options: [status(), feed(), play(), sleep(), attack(), quit()])\n")
        if choice == "quit()":
            print(f"\nGoodbye! {monster.monster_name} will miss you!")
            break  # Exit the loop and end the program
        elif choice == "attack()":
            other_name = input("Enter the name of the monster you want to attack: ")
            other_monster = CombatMonster(other_name)
            monster.attack(other_monster)
        else:
            try:
                loop = eval(f"monster.{choice}")
            except:
                print('You have typed an error execution, please try again.')

if __name__ == '__main__':
    main()
```
