```python
## P1
class Pet:
    def __init__(self, name):
        self.pet_name = name
        self.hunger = 5  # Scale of 1-10, where 10 is very hungry
        self.energy = 5  # Scale of 1-10, where 10 is full of energy
        self.happiness = 5  # Scale of 1-10, where 10 is very happy

    def status(self, evolution=0):
        print(f"\n{self.pet_name}'s status: Hunger = {self.hunger}, Energy = {self.energy}, Happiness = {self.happiness}")
        if evolution and self.hunger >= 10 and self.energy >= 10 and self.happiness >= 10:
            print(f"\n{self.pet_name} has successfully evolved.")
            print(f"1. You need to upgrade your code yourself to complete the subsequent evolution design or even battles.")
            print(f"2. You can design or upload a character for your companions.")
            print(f"For more information, please contact your teacher.")
            return False
        else:
            return True

    def feed(self):
        if self.hunger < 9:
            self.hunger += 2  # Feeding decreases hunger
            self.happiness += 1
            print(f"\nYou fed {self.pet_name}. Yum!")
        else:
            print(f"\n{self.pet_name} is not very hungry right now.")
        self.status()
        return True

    def play(self):
        if self.energy > 2:
            if self.happiness < 9:
                self.happiness += 2  # Playing increases happiness
            else:
                self.happiness = 10
            self.energy -= 2  # Playing decreases energy
            self.hunger -= 1
            print(f"\nYou played with {self.pet_name}. Fun!")
        else:
            print(f"\n{self.pet_name} is too tired to play.")
        self.status()
        return True

    def sleep(self):
        if self.energy < 8:
            self.energy += 3  # Sleeping increases energy
            self.hunger -= 1
            self.happiness -= 1
            print(f"\n{self.pet_name} is sleeping. Zzz...")
        else:
            print(f"\n{self.pet_name} is not very tired, {self.pet_name} do not want to sleep.")
        self.status()
        return True

def main():
    loop = True
    while loop:
        choice = input(f"\nWhat would you like to do with {name}? \n(Hint: You can type in one of your options: [status(),feed(),play(),sleep(),quit()])\n")
        if choice == "quit()":
            print(f"\nGoodbye! {name} will miss you!")
            break  # Exit the loop and end the program
        else:
            try:
                loop = eval(f"pet.{choice}")
            except:
                print('You have typed an error execution, please try again.')

name = input('Please give your pet a name: ')
pet = Pet(name)

if __name__ == '__main__':
    main()
```

```python
## P2
class TicTacToe:
    def __init__(self):
        self.board = [[" " for _ in range(3)] for _ in range(3)]
        self.current_player = "X"

    def print_board(self):
        separator = " +---+---+---+"
        print("   0   1   2")
        print(separator)
        for idx, row in enumerate(self.board):
            print(f"{idx}| {' | '.join(row)} |")
            print(separator)
        print()

    def check_winner(self, player):
        for i in range(3):
            if all(self.board[i][j] == player for j in range(3)) or \
               all(self.board[j][i] == player for j in range(3)):
                return True
        if all(self.board[i][i] == player for i in range(3)) or \
           all(self.board[i][2 - i] == player for i in range(3)):
            return True
        return False

    def play(self):
        while True:
            self.print_board()
            move = input(f"Player {self.current_player}, enter row and column (e.g., '1 2'): ")
            try:
                row, col = map(int, move.split())
            except ValueError:
                print("Invalid input. Please enter two integers separated by a space.")
                continue
            if not (0 <= row <= 2 and 0 <= col <= 2):
                print("Invalid input. Row and column must be between 0 and 2.")
                continue
            if self.board[row][col] != " ":
                print("That position is already taken!")
                continue
            self.board[row][col] = self.current_player
            if self.check_winner(self.current_player):
                self.print_board()
                print(f"Player {self.current_player} wins!")
                break
            if all(self.board[i][j] != " " for i in range(3) for j in range(3)):
                self.print_board()
                print("It's a tie!")
                break
            self.current_player = "O" if self.current_player == "X" else "X"

if __name__ == "__main__":
    game = TicTacToe()
    game.play()
```

```python
## P3
class MiniATMSimulator:
    def __init__(self, account_number, pin, starting_balance):
        self.account_number = account_number
        self.pin = pin
        self.balance = starting_balance

    def login(self):
        user_account = input("Enter your account number: ")
        user_pin = input("Enter your PIN: ")
        if user_account == self.account_number and user_pin == self.pin:
            return True
        else:
            return False

    def deposit(self):
        amount = float(input("Enter amount to deposit: "))
        self.balance += amount
        print(f"${amount} deposited successfully.")

    def withdraw(self):
        amount = float(input("Enter amount to withdraw: "))
        if amount <= self.balance:
            self.balance -= amount
            print(f"${amount} withdrawn successfully.")
        else:
            print("Insufficient balance.")

    def check_balance(self):
        print(f"Your current balance is: ${self.balance}")

    def main(self):
        if self.login():
            print("Login successful!")
            while True:
                print("\n1. Deposit")
                print("2. Withdraw")
                print("3. Check Balance")
                print("4. Exit")
                choice = input("Choose an option (1-4): ")

                if choice == "1":
                    self.deposit()
                elif choice == "2":
                    self.withdraw()
                elif choice == "3":
                    self.check_balance()
                elif choice == "4":
                    print("Thank you for using the Mini ATM Simulator. Goodbye!")
                    break
                else:
                    print("Invalid choice. Please enter a number between 1 and 4.")
        else:
            print("Invalid account number or PIN.")

if __name__ == "__main__":
    atm = MiniATMSimulator("123456", "1234", 1000)
    atm.main()
```

```python
## P4
class GridGame:
    def __init__(self, size=5):
        self.size = size
        self.player_x, self.player_y = 0, 0  # Starting position of the player

    def print_grid(self):
        for x in range(self.size):
            for y in range(self.size):
                if x == self.player_x and y == self.player_y:
                    print("P ", end="")  # Print P for Player's current position
                else:
                    print(". ", end="")  # Print . for empty space
            print()  # New line at the end of each row

    def play_game(self):
        while True:
            print("\nCurrent map:")
            self.print_grid()  # Show the grid

            move = input("Move (WASD): ").upper()
            if move == "W" and self.player_x > 0:
                self.player_x -= 1
            elif move == "S" and self.player_x < self.size - 1:
                self.player_x += 1
            elif move == "A" and self.player_y > 0:
                self.player_y -= 1
            elif move == "D" and self.player_y < self.size - 1:
                self.player_y += 1
            else:
                print("Invalid move or boundary reached. Please try again.")

if __name__ == "__main__":
    print("Navigate inside the grid. Use WASD keys to move: W (up), A (left), S (down), D (right).")
    game = GridGame()
    game.play_game()
```
